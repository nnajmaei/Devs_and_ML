# Path Planning with Numpy Example

```python
from path_planning_numpy import *
```

### Setting the Environment Parameters

```python
start=np.array([0,0])
end=np.array([10,10])
min_size=0.5
max_size=1
per_obs_coef=0.5
step=0.25
```

### Creating Randomized Obstacles.
The first number is the number of obstacles created about the line connecting start and end. and the second number is the number of obstacles created in the environment.

```python
obstacles = create_random_obstacles(4,50,start,end,min_size,max_size,per_obs_coef)
print("The number of random obstacles created:",len(obstacles))
```

### Running the Path Planning Algorithms
Finds all the possible paths from start to end

```python
paths = finding_all_routes(start,end,step,obstacles)
```

### Finding the shortest path based on distance travelled

```python
shortest_p = shortest_path(paths)
print("The shortest path has a length of: {:.4f}".format(shortest_p.calculate_distance()))
```

### Plotting the results
(all paths + optimal path)

```python
figure, axes = plt.subplots()
plot_path(paths,start,end,obstacles,axes)
figure2, axes2 = plt.subplots()
plot_path(shortest_p,start,end,obstacles,axes2)
plt.show()
```
