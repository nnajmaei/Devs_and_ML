# Clustering Implementation
## Using k-mean algorithm

```python
from clustering import *
```

### Generating random clusters

We create a random number of clusters, consisted of random number of points in a space

```python
x_max = 15
y_max = 15
z_max = 15
radius = 2.5
min_groups = 6
max_groups = 12
k = 2

k_init = 5
k_max = 10
max_tries = 3

set_of_points = []
set_of_clusters = []
ks = []
inits = []
```

## The mean-k algorithm
Performed for k=2 to 10 to find the best clustering

```python
for i in range(0, 20):
    radius = random.uniform(1, 3)
    points = gerenate_random_vectors(
        x_max, y_max, z_max, radius, min_groups, max_groups
    )
    clusters, k, init = clustering_k(points, k_init, k_max, max_tries)

    set_of_points.append(points)
    set_of_clusters.append(clusters)
    ks.append(k)
    inits.append(init)
    print("k " + str(k) + ", init: " + str(init))

print("FINISHED")
```

```python
for clusters in set_of_clusters:
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    for cluster in clusters:
        plot_points(cluster.points_list, plt, ax)

    ax.set_aspect("equal", adjustable="box")
    plt.show()
```
