## Vectors - 2

$$V=\{v_1,v_2,...,v_n\}$$

#### Linear Combination
$$ w = \lambda_1 v_1+ ... + \lambda_n v_n$$

#### Linear Independence
A set of vectors V is said to be linearly dependent, if ther exists a set of coefficients for which:
$$ 0 = \lambda_1 v_1+ ... + \lambda_n v_n,    \lambda_i \neq 0 \in \mathbb{R} $$

#### Subpace, Span, Basis
The infinite options of linear combination of a set of vectors, will creaate a subspace. The dimension of the subspace is the number of linearly independent vectors in the set that creates the subspace.
A set of vectors <u> spans </u> and the result is a <u> subspace </u>. And span is a mechanism of creating a subspace.

A set of vectors V creates a <u> basis </u> of a subspace if (1) the V spans the subspace, and (2) the vectors in V are linearly indep.

```python
# A for loop for creating the linear combination
scalars = [3,4,5]
v1=np.array([[1,2]])
v2=np.array([[2,2]])
v3=np.array([[3,2]])

vs=[v1,v2,v3]
lincombo=np.zeros(np.shape(v1))

for s,v in zip(scalars,vs):
    lincombo += s*v
    
print(lincombo)
```

### Correlation
if $x$ and $y$ are two vectors which are <u>demeaned</u>, then the correlation between two vectors are between -1 and 1 and is formulated as:
$$\rho = \frac{x^Ty}{||x||.||y||}$$

This is similar to the <u>cosine similarity</u> but with a difference that this does not involve a normalized vectors. $cos(\theta_{x,y})=\frac{x^Ty}{||x||||y||}$

```python
#Excercise 4-1
def corandsine(x,y):

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    x_m=(x-mean_x)    
    y_m=(y-mean_y)
    
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    norm_xm = np.linalg.norm(x_m)
    norm_ym = np.linalg.norm(y_m)
    
    cor = np.dot(x_m,y_m)/(norm_xm*norm_ym)
    cos = np.dot(x,y)/(norm_x*norm_y)
    
    return cor,cos

a = np.random.randn(3)
b = np.random.randn(3)
cor, cos = corandsine(a,b)
print(f'a= {a}')
print(f'b= {b}')
print(f'Correlation between a and b is: {np.round(cor,4)}')
print(f'Correlation between a and b is: {np.round(np.corrcoef(a,b)[0,1],4)} (using numpuy function)')
print(f'Correlation between a and b is: {np.round(pearsonr(a,b)[0],4)} (using scipy function)')
print(f'Cosine Sim between a and b is: {np.round(cos,5)}')
```

```python
a = np.arange(4,dtype=float)
offsets = np.arange(-50,51)

results = np.zeros((len(offsets),2))

for i in range(len(offsets)):
    results[i,:] = corandsine(a,a+offsets[i])

fig = plt.figure(figsize=(4,2))
h = plt.plot(offsets,results)
h[0].set_color('k')
h[0].set_marker('o')
h[1].set_color([.7,.7,.7])
h[1].set_marker('s')

plt.xlabel('Mean offset')
plt.ylabel('r or c')
plt.legend(['Pearson','Cosine sim.'])
plt.show()
```

### Filtering and Feature Detection


```python
signal = np.zeros(30)
signal[10:20]=1


fig = plt.figure(figsize=(3,3))

plt.plot(signal,marker='o')


kernel = np.array([-1,1])
#kernel = np.array([1,-1])
#kernel = np.array([-1,2])
#kernel = np.array([0,-1,1,0])
filtered_signal = np.zeros(signal.shape)


for t in range(1,len(signal)-len(kernel)+1):
    filtered_signal[t]=np.dot(kernel,signal[t-1:t+len(kernel)-1])

plt.plot(filtered_signal,marker='x')

plt.show()
```

###### ============================
### K-Mean Clustering

1- Initialize k clusters with random centroid (represetative vectos)
2- Assign each point ot the cluster with closest centroid
3- Update centroids to be the average of the points in the cluster
4- repear 2 and 3 until the objective function is converged

```python
#creating random points
v1=np.array([2,5,1])
v2=np.array([0,2,2])
v3=np.array([3,4,5])
xlim = [-3,3]

num_points = 200
scalars = np.random.uniform(low=xlim[0],high=xlim[1],size=(num_points,3))
data=np.zeros((num_points,3))

for i in range(len(scalars)):
    data[i,:]=scalars[i,0]*v1+scalars[i,1]*v2+scalars[i,2]*v3

fig = plt.figure(figsize=(3,3))
ax=plt.axes(projection="3d")

ax.scatter3D(data[:,0],data[:,1],data[:,2])
plt.show()


print(f'The shape of data set is {data.shape}\n')

#randomizing the centroids
k=4
ridx = np.random.choice(range(len(data)),(k),replace=False)
print(f'The shape of ridx set is {ridx.shape}\n')

centroids = data[ridx,:]
centroids_temp=np.zeros(centroids.shape)

print(f'The shape of centroids set is {centroids.shape}')
print(f'The random centroids are:\n{centroids}\n')

while np.sum(abs(centroids-centroids_temp))>0.000001:
    #calculating distance to centroids
    centroids_temp=centroids.copy()
    dists=np.zeros((data.shape[0],k))
    
    #use of broadcasting to calculate distance to the centoirds
    for ci in range(k):
        dists[:,ci]=np.sum((data-centroids[ci,:])**2,axis=1)

    #finding the closests centroid
    groupidx = np.argmin(dists,axis=1)
    
    
    for ki in range(k):
        centroids[ki,:]=[np.mean(data[groupidx==ki,0]),np.mean(data[groupidx==ki,1]),np.mean(data[groupidx==ki,2])]

print(f'The shape of dists set is {dists.shape}\n')
print(f'The shape of groupidx set is {groupidx.shape}\n')
print(f'The finalized centroids are:\n{centroids}\n')

fig = plt.figure(figsize=(3,3))
ax=plt.axes(projection="3d")

for ki in range(k):
    ax.scatter3D(data[groupidx==ki,0],data[groupidx==ki,1],data[groupidx==ki,2],marker='o')

ax.scatter3D(centroids[:,0],centroids[:,1],centroids[:,2],marker='^')
plt.show()

```
