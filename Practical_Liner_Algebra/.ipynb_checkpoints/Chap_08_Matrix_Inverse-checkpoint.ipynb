## Matrix Inverse

- $A^{-1}$ is the inverse of $A$ and $A^{-1}A=I$
- $(AB)^{-1}=B^{-1}A^{-1}$
- $Ax=b \implies x=A^{-1}b$ solves the equation
- Matrix inverse is unique. But for a singular matrix, the pseudoinverse is not unique

#### Types
- <u>Full Inverse</u>: for square and full-rank matrices. $A^{-1}A=AA^{-1}=I$ 
- <u>One-sided Inverse</u>: for rectangular and full-rank matrices: for a tall Matrix (full column rank), left sided $LT=I, TL\neq I$. For a wide mattix (full row rank), right sided $WR=I, RW\neq I$
- <u>Pseudoinverse</u>: for any matrix with any shape and rank, there exist a matrix that transforms it to a matrix close to indentity. In the two aboves cases, the pseudoinverse is equal to full or one-sided inverse.

#### Calculting Matrix Iverse of square full-rank $A$
- Calculate Minors Matrix $M$, where $M_{i,j}$  is the determinant of matrix A, when the i-th row and j-th column is deleted.
- Calculate Grid Matrix $G$, where $G_{i,j}=-1^{i+j}$
- Calculate cofactor matrix $C=M \times G$, by hadamard multilication
- Calculate adjugate or inverse matrix by $A^{-1}=\frac{C^T}{det(A)}$

OR

- np.linalg.inv()

#### One-sided inverse of $T$ or $W$
- $LT = I \implies L=(T^TT)^{-1}T^T$
- $WR = I \implies R=W^T(WW^T)^{-1}$

#### Moore-Penrose Pseuoinverse
np.linalg.pinv()

```python
#exer 3: full-iverse
m=5
A=np.random.randn(m,m)

#function to obtain inverse
def matrix_inverse(A):
    A_inv=np.zeros(A.shape)
    #calculating minor matrices and grod matrices
    minorMat = np.zeros(A.shape)
    gridMat  = np.zeros(A.shape)
        
    for i in range(minorMat.shape[0]):
        for j in range(minorMat.shape[1]):
            gridMat[i,j]=(-1)**(i+j)
            
            A_reduced = np.delete(A,i,axis=0)
            A_reduced = np.delete(A_reduced,j,axis=1)
            
            minorMat[i,j]=np.linalg.det(A_reduced)
                        
    # Cofactor matrix
    cofactMat = np.multiply(minorMat,gridMat)
    
    
    # Adjugate or inverse matrix
    A_inv = cofactMat.T/np.linalg.det(A)
    
    return A_inv, minorMat, cofactMat, gridMat

A_inv1, minorMat, cofactMat, gridMat=matrix_inverse(A)

A_inv2=np.linalg.inv(A)

fig,axs = plt.subplots(2,3,figsize=(6,3))

axs[0,0].imshow(minorMat,cmap='gray')
axs[0,0].set_title("Minors Mat")

axs[0,1].imshow(gridMat,cmap='gray')
axs[0,1].set_title("Grid Mat")

axs[0,2].imshow(cofactMat,cmap='gray')
axs[0,2].set_title("Cofactor Mat")

axs[1,0].imshow(A_inv1,cmap='gray')
axs[1,0].set_title("Adjugate Mat")

axs[1,1].imshow(A_inv2,cmap='gray')
axs[1,1].set_title("NumPy Mat")

axs[1,2].imshow(A_inv2@A,cmap='gray')
axs[1,2].set_title("A@Ainv")
```

```python
#Left-iverse
m=10
n=2
A=np.random.randn(m,n)

#function to obtain inverse
def matrix_left_inverse(A):
    return np.linalg.inv(A.T@A)@A.T


A_linv=matrix_left_inverse(A)


fig,axs = plt.subplots(2,2,figsize=(6,3))

axs[0,0].imshow(A,cmap='gray')
axs[0,0].set_title("Tall Mat")

axs[0,1].imshow(A_linv,cmap='gray')
axs[0,1].set_title("Left Inv Mat")

axs[1,0].imshow(A_linv@A,cmap='gray')
axs[1,0].set_title("A_inv@A")

axs[1,1].imshow(A@A_linv,cmap='gray')
axs[1,1].set_title("A@A_inv")
```

```python
#Right-iverse
m=2
n=10
A=np.random.randn(m,n)

#function to obtain inverse
def matrix_left_inverse(A):
    return A.T@np.linalg.inv(A@A.T)


A_linv=matrix_left_inverse(A)


fig,axs = plt.subplots(2,2,figsize=(6,3))

axs[0,0].imshow(A,cmap='gray')
axs[0,0].set_title("Tall Mat")

axs[0,1].imshow(A_linv,cmap='gray')
axs[0,1].set_title("Left Inv Mat")

axs[1,0].imshow(A_linv@A,cmap='gray')
axs[1,0].set_title("A_inv@A")

axs[1,1].imshow(A@A_linv,cmap='gray')
axs[1,1].set_title("A@A_inv")
```

###### ============================
### Numerical Stability of Inverse

As the range of values in a matrix becomes larger, calculating the determinants and inverse of matrices becomes numerically unstable.

Overall matrices with low range of variation in their values, have higher stability. And with large range, lower stability.

<u>Condition number of a matrix</u> is ratio of the largest to the smallest singular value. It is a measure of the spread of values in a matrix.
np.linalg.cond()

One way to avoid instability is that the algorithms use a decoomposed version of the matrix, which gives it higher stability.

###### ============================
### Gerometrical Interpretion

Inverse matrix, undo the transformation a matrix applied to a vector.


```python
# Exer 7 Transformation matrix
T = np.array([ 
              [1,.5],
              [0,.5]
            ])

# Compute its inverse
Ti = np.linalg.inv(T)


# define the set of points (a circle)
theta = np.linspace(0,2*np.pi-2*np.pi/20,20)
origPoints = np.vstack( (np.cos(theta),np.sin(theta)) )

# apply transformation
transformedPoints = T @ origPoints

# undo the transformation via the inverse of the transform
backTransformed   = Ti @ transformedPoints


# plot the points
plt.figure(figsize=(3,3))
plt.plot(origPoints[0,:],origPoints[1,:],'ko',label='Original')
plt.plot(transformedPoints[0,:],transformedPoints[1,:],'s',color=[.7,.7,.7],label='Transformed')
plt.plot(backTransformed[0,:],backTransformed[1,:],'rx',markersize=15,label='Inverse-transformed')

plt.axis('square')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.legend(fontsize=5)
plt.show()
```

```python
#Exer 8 and 9 - Numerical stability and Hilber Matrix
#Method double for loop
def hilbert_mat(n):
    hilbertMat=np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            hilbertMat[i,j]=1/(i+j+1)
    return hilbertMat

# The double for-loop above is a direct implementation of the math.
# The function below gives the same result but without the loops. using outer product and broadcasting
k=10

def hilbmat(k):
    k = np.arange(1,k+1).reshape(1,-1) # reshape to a row vector (instead of a 1D array)
    return 1 / (k+k.T-1) # outer product and element-wise division

hilbert_1=hilbert_mat(k) #from double for loops
hilbert_2=hilbmat(k) #from function above
hilbert_3=hilbert(k) #from scipy function hilbert imported

print(f'distance between the for loops and outer is: {np.linalg.norm(hilbert_1-hilbert_2)}')
print(f'distance between the for loops and scipy is: {np.linalg.norm(hilbert_1-hilbert_3)}')
```

```python
#Exer 8 and 9 - Numerical stability and Hilber Matrix
k=10

H=hilbert(10)
Hi = np.linalg.inv(H)


fig,axs = plt.subplots(1,3,figsize=(9,3))
h = [0,0,0]

# the matrix
h[0] = axs[0].imshow(H,cmap='gray')
axs[0].set_title('Hilbert matrix')

# its inverse
h[1] = axs[1].imshow(Hi,cmap='gray')
axs[1].set_title('Inv(hilbert)')

# their product
h[2] = axs[2].imshow(H@Hi,cmap='gray')
axs[2].set_title('Their product')


for i in range(2):
  fig.colorbar(h[i],ax=axs[i],fraction=.045)
  axs[i].set_xticks([])
  axs[i].set_yticks([])

plt.tight_layout()
plt.show()
```

```python
matSizes = np.arange(3,13)

identityError = np.zeros((len(matSizes),2))
condNumbers   = np.zeros((len(matSizes),2))


for i,k in enumerate(matSizes):
    
    ### for the Hilbert matrix
    H   = hilbmat(k)       # the matrix
    Hi  = np.linalg.inv(H) # its inverse
    HHi = H@Hi             # should be identity
    err = HHi - np.eye(k)  # difference from true identity
    identityError[i,0] = np.sqrt(np.sum(err**2))  # Euclidean distance
    condNumbers[i,0] = np.linalg.cond(H) # condition number
    

    ### repeat for a random matrix
    H = np.random.randn(k,k) # the matrix
    Hi  = np.linalg.inv(H)   # its inverse
    HHi = H@Hi               # should be identity
    err = HHi - np.eye(k)    # difference from true identity
    identityError[i,1] = np.sqrt(np.sum(err**2))  # Euclidean distance
    condNumbers[i,1] = np.linalg.cond(H) # condition number
    


# now plot
fig,axs = plt.subplots(1,2,figsize=(6,3))

## plot the Euclidean distance to the identity matrix
h = axs[0].plot(matSizes,np.log(identityError),'s-',markersize=12)
h[0].set_color('k') # adjust the individual line colors and shapes
h[0].set_marker('o')
h[1].set_color('gray')

axs[0].legend(['Hilbert','Random'])
axs[0].set_xlabel('Matrix size')
axs[0].set_ylabel('Log Euclidan distance')
axs[0].set_title('Distance to identity matrix')



## plot the condition numbers
h = axs[1].plot(matSizes,np.log(condNumbers),'s-',markersize=12)
h[0].set_color('k') # adjust the individual line colors and shapes
h[0].set_marker('o')
h[1].set_color('gray')

axs[1].legend(['Hilbert','Random'])
axs[1].set_xlabel('Matrix size')
axs[1].set_ylabel('Log Kappa')
axs[1].set_title('Matrix condition number')

plt.show()
```
