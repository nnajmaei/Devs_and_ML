```python
#Exer 3

C=np.arange(100).reshape(10,10)


C_1=C[5:10,5:10]
C_2=C[5:10,0:5]
C_3=C[0:5,5:10]
C_4=C[0:5,0:5]


C_new=np.vstack((np.hstack((C_1,C_2)),np.hstack((C_3,C_4))))


_,axs = plt.subplots(1,2,figsize=(6,3))
axs[0].imshow(C,cmap='gray',origin='upper',vmin=0,vmax=np.max(C))
axs[0].set_title('Original matrix')
for (j,i),num in np.ndenumerate(C):
    axs[0].text(i,j,num,color=[0.5,0.5,0.5],ha='center',va='center')
    
axs[1].imshow(C_new,cmap='gray',origin='upper',vmin=0,vmax=np.max(C))
axs[1].set_title('Submatrix')
for (j,i),num in np.ndenumerate(C_new):
    axs[1].text(i,j,num,color=[0.5,0.5,0.5],ha='center',va='center')

```

```python
A=np.ones((4,4))
print(A)
print("\n")

B=np.diag([1,4,9,16])
print(B)
print("\n")

C=np.sqrt(B)
print(C)
print("\n")

print(B@A)
print("\n")

print(A@B)
print("\n")

print(C@A@C)
```

###### ============================
### Matrix Math

##### Shifting
- Addition of a scalar to the diagonal of a marix $\matrix{A}+\lambda \matrix{I}$

##### Matrix Multiplication
- Matrix multiplication is only valid when the inner dimension of the two matrix are equal. The outer dimensions are the dimension of the result matrix. (MxN)*(NxK)=(MxK)

- Each element i,j of the multiple matrix is the dot product of row i of the left matrix and column j of the right matrix.

- Dot product of two vectors is a number that encode the relationship of the two vectors. HENCE, matrix multiplican is a matrix that stores all the pairwise linear relationships between rows of the left matrix and columns of the right matrix.

- We can also say it is an organized collection of mapping between row-columns of two matrix

##### Matrix Vector Multiplication
- Function of this mult is that Matrix house a transformation that when applied to a vector, can rotate and stretch the vector

##### Transpose
$$(AB)^T=B^TA^T$$

```python
print(f'A=\n{A}')
B= np.random.randint(1,4,size=(3,3))
print(f'B=\n{B}\n======')

print(f'Addition: A + B =\n{A+B}\n======')

print(f'Scalar Addition: A + 3 =\n{A+3}\n======')

print(f'Shifting: A + 3I =\n{A+3*np.eye(len(A))}\n======')

print(f'Shifting: 3 * A =\n{3*A}\n======')

print(f'Hadamard Multiplication (element-wise mult) v1: A*B =\n{A*B}\n======')
print(f'Hadamard Multiplication (element-wise mult) v2: A*B =\n{np.multiply(A,B)}\n======')

print(f'Standard Matrix Multiplication: A@B =\n{A@B}\nAlso using dot product:')

#Matrix multiplication A*B is simply dot product of rows of A and columns of B
M = np.zeros((A.shape[0],B.shape[1]))

for i in range(A.shape[0]):
    for j in range(B.shape[1]):
        M[i,j]=int(np.dot(A[i,:],B[:,j]))

print(f'{M}\n======')

print(f'Transpose: A^T =\n{A.T}\n======')

print(f'(AB)^T=B^T.A^T =\n{(A@B).T}=\n{B.T@A.T}\n======')
```

###### ============================
### Symmetric Matrix and Creating Sym matrix from NonSym

A symmetric matrix is always square and $A^T=A$.

<u>Multiplicative Method</u>: For any matrix $\matrix{A}$, $\matrix{A}^T\matrix{A}$ and $\matrix{A}\matrix{A}^T$ are symmetric.




```python
A=np.random.randint(1,5,size=(5,6))

print(f'Original Matrix A {A.shape}:\n{A}')

print(f'A^T.A is symm {(A.T@A).shape}:\n{A.T@A}')
print(f'A.A^T is symm {(A@A.T).shape}:\n{A@A.T}')
```
